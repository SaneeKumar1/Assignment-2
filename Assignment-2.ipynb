Q1. How do you comment code in Python? What are the different types of comments?

Ans:- In Python, we can add comments to our code to provide explanations or make notes that are not executed as part of the program. There are two types of comments in Python.

(1) single-line comments,
(2) multi-line comments

(1) Single-line comments:- Single-line comments are used to comment on a single line of code. They start with the # symbol and continue until the end of the line.

(2) Multi-line comments: Multi-line comments, also known as block comments, are used to comment on multiple lines of code. We can use triple quotes (""" """ or''' ''') to create multi-line strings that are not assigned to any variable. These strings act as comments and are ignored by the interpreter.

Examples:- 

## This is a single-line comment

"""
This is a multi-line comment.
It can span across multiple lines.
"""

Q4. How do you write and execute a Python script from the command line?

Ans:- To write and execute a Python script from the command line, follow these steps:

1. Open a text editor and create a new file with a .py extension, such as script.py.

2. Write your Python code in the file and save it.

3. Open a command prompt or terminal.

4. Navigate to the directory where the Python script is located using the cd command.

5. Run the Python script by entering python script.py in the command prompt or terminal.

6. Press Enter to execute the command.

The Python interpreter will read and execute the code in our script file. Any output generated by the script will be displayed in the command prompt or terminal window.

Q6. What is a complex number in mathematics, and how is it represented in Python?

Ans:- In mathematics, a complex number is a number that consists of a real part and an imaginary part. It is written as "a + bi," where "a" represents the real part and "b" represents the imaginary part.

In Python, complex numbers are represented using the built-in complex type. The real and imaginary parts are specified as arguments to the complex constructor or by using the 'j' suffix to denote the imaginary part. 

For example:

```python
# Creating a complex number using the complex constructor
z1 = complex(3, 4)  # 3 + 4i

# Creating a complex number using the j suffix
z2 = 2 + 5j       # 2 + 5i
```
